Logic for extracting Subject, Object and relation from the sentence

```python
def get_entities(sent):
    ent1 = ""
    ent2 = ""

    prv_tok_dep = ""
    prv_tok_text = ""

    prefix = ""
    relation = ""
    for tok in sentence1:
        # for tok in item:
            if tok.dep_ != "punct":
                if tok.dep_ == "compound":
                    prefix = tok.text
                if tok.dep_.endswith("acl") == True:
                    relation = tok.text
                # elif tok.dep_.endswith("ROOT") == True:
                #     relation = tok.text

                
                if tok.dep_.endswith("pobj") == True:
                    ent2 = prefix +" "+ tok.text
                if tok.dep_.find("subj") == True:
                    ent1 = relation +" "+ prefix + " "+ tok.text
                    prefix = ""
                    relation = ""

    return [ent1.strip(),relation.strip(), ent2.strip()]
             
```
